
# SISTEMAS BINARIOS

# Introducción

El estudio de sistemas binarios es una herramienta importante para determinar masas y radios estelares.  
Un sistema binario físico es un conjunto de dos estrellas que orbitan alrededor del centro de masa debido a la interacción gravitatoria entre ellas. La clasificación de estos sistemas dependerá de la forma en que se detecten.

<img src="esquemasistemaestelar.jpeg" width="500" />


## Tipos de sistemas binarios

- **Binarias visuales**: las dos componentes están lo suficientemente separadas (órbitas grandes) como para obtener información de ambas a través del telescopio.
- **Binarias espectroscópicas**: las componentes tienen variaciones de velocidad radial “considerables” en períodos cortos (órbitas pequeñas).
- **Binarias eclipsantes**: el plano orbital del sistema es casi perpendicular al plano del cielo y cada componente del par pasará periódicamente entre el observador y la otra componente, produciendo un eclipse y disminuyendo la luz que llega del sistema.

En este trabajo nos centraremos en las binarias espectroscópicas, aquellas que se descubren por los corrimientos periódicos de las líneas del espectro. Pueden verse los dos espectros (SB2) o, si la estrella secundaria es mucho más débil que la primaria, se observa sólo un espectro (SB1).  
El movimiento orbital genera un corrimiento de las líneas espectrales características de cada estrella por efecto Doppler. Siendo este corrimiento máximo cuando las estrellas se mueven en la dirección de la visual y mínimo o nulo cuando "pasan" por la dirección perpendicular como se observa en la siguiente figura.

<img src="esquema_espectro.png" width="500" />


Con la obtención de los espectros a lo largo de un tiempo igual o mayor al período orbital se construyen las curvas de velocidad radial de las componentes del sistema.

<img src="esquema_velr.jpg" width="500" />


## Elementos orbitales

La proyección en el plano del cielo de la órbita de una estrella alrededor del centro de masa es una elipse caracterizada por los elementos orbitales:

- **e**: excentricidad de la órbita
- **a**: semieje mayor de la órbita
- **i**: inclinación del plano orbital respecto al plano del cielo.
- **ω**: Argumento del periastro.

<img src="proyeccion_elipse.png" width="500" />


Donde $r$ es el radio-vector que describe paramétricamente la órbita en función del tiempo, completando la misma en un período $P$, y $\nu$ es la anomalía verdadera.

$$
r = \frac{a(1-e^2)}{1+e \cos(\nu)}
$$

En este esquema la velocidad radial $V_R$ se describe como:

$$
V_R = V_0 + \frac{dz}{dt}
$$

Donde $V_0$ es la velocidad baricéntrica del sistema. Luego, utilizando la segunda ley de Kepler y consideraciones geométricas se encuentra una expresión para $\frac{dz}{dt}$ en función de la variación del radio-vector $r(t)$.  
Finalmente se obtiene una expresión para la velocidad radial en cada instante en función de los parámetros orbitales.

$$
V_R = V_0 + K[\cos(\omega) + \cos(\omega + \nu)]
$$

con $K = \frac{a \sin(i)}{\sqrt{(1-e^2)}}$.  
$K$ se denomina semiamplitud de la órbita de velocidad radial, ya que de la ecuación de $V_R$ (3) se observa que el máximo se da cuando $\cos(\omega + \nu) = 1$ y el mínimo cuando $\cos(\omega + \nu) = -1$, por lo tanto, $V_{Rmax} - V_{Rmin} = 2K$.



La siguiente figura muestra varias curvas de velocidad radial vs fase para distintos valores de excentricidad y de $\omega$. $\gamma$ representa el valor de la velocidad baricéntrica $V_0$.

<img src="curvasvr.png" width="500" />


Conocer los parámetros orbitales de las componentes del sistema es importante porque permite obtener información sobre las masas estelares.  
En el caso SB1 (un sólo espectro) podemos conocer la función de masa $f(M)$.

$$
f(M) = \frac{(a_1 \sin(i))^3}{P^2}
$$

Y en el caso de obtener ambos espectros (SB2), a partir del cociente de masas $q$ definido como:

$$
q = \frac{M_2}{M_1} = \frac{K_1}{K_2}
$$

y utilizando la 3era Ley de Kepler se pueden obtener las masas mínimas de ambas componentes.

$$
M_1 \sin^3(i) = \frac{a_2 \sin(i)}{P^2} (a_1 \sin(i) + a_2 \sin(i))^2
$$

$$
M_2 \sin^3(i) = \frac{a_1 \sin(i)}{P^2} (a_1 \sin(i) + a_2 \sin(i))^2
$$

donde los subíndices 1 y 2 indican las componentes primaria y secundaria respectivamente.

## Objetivo

El objetivo de esta práctica es determinar las masas mínimas de las componentes del sistema binario $V501$ Mon a través del método de Lehmann-Filhés. Luego comparar los resultados con los obtenidos en el estudio de Torres et al. (2015).




El método de Lehmann-Filhés consiste en obtener el conjunto de parámetros orbitales que mejor representan las observaciones mediante iteraciones conocidas como "correcciones diferenciales".  
El cambio de velocidad radial ($\Delta V_r$) frente a incrementos en los parámetros $K$, $e$, $\omega$, $t_0$, $P$ y $V_0$ se puede desarrollar en una serie de Taylor a primer orden.

$$
\Delta V_r = \frac{\partial V_r}{\partial K} \Delta K + \frac{\partial V_r}{\partial e} \Delta e + \frac{\partial V_r}{\partial \omega} \Delta \omega + \Delta V_0
$$

En este caso se considera fijo $P$ y $t_0$. Las expresiones para las derivadas parciales son:

$$
\frac{\partial V_r}{\partial K} = \cos(\nu + \omega) + e \cos(\omega)
$$

$$
\frac{\partial V_r}{\partial e} = K \left( \cos(\omega) - \frac{\sin(\nu + \omega) \sin(\nu)(2 + e \cos(\nu))}{(1 - e^2)} \right)
$$

$$
\frac{\partial V_r}{\partial \omega} = - K (\sin(\nu + \omega) + e \sin(\omega))
$$

Para obtener los parámetros de mejor ajuste, en primer lugar se realizan estimaciones iniciales a partir de la observación de la curva de $V_r$ vs fase $\phi$ calculada con los datos. Con estos parámetros se calculan las velocidades radiales en forma teórica a partir de la ecuación \ref{v_r_fin} y los valores numéricos de las derivadas parciales $\frac{\partial V_r}{\partial e}$, $\frac{\partial V_r}{\partial K}$ y $\frac{\partial V_r}{\partial \omega}$.

Luego, los incrementos $\Delta e$, $\Delta K$, $\Delta \omega$ y $\Delta V_0$ se calculan ajustando por mínimos cuadrados la diferencia entre los valores observados ($O$) de velocidad radial y los calculados teóricamente ($C$), conocido como $ \text{residuo} = (O - C) $.

Una vez obtenidos los incrementos, se vuelven a calcular las velocidades radiales y las derivadas parciales con los parámetros corregidos:

$$
k' = k + \Delta K
$$

$$
e' = e + \Delta e
$$

$$
\omega' = \omega + \Delta \omega
$$

$$
V_0' = V_0 + \Delta V_0
$$

El proceso se repite hasta obtener la diferencia $(O - C)$ que se desee.

 
A su vez, la fase $\phi$ se relaciona con la anomalía verdadera a través de la ecuación de Kepler:

$$
E - e \sin(E) = 2\pi \phi
$$

y la anomalía verdadera $\nu$ con la anomalía excéntrica $E$ a través de:

$$
\tan\left(\frac{\nu}{2}\right) = \left(\frac{1+e}{1-e}\right)^{\frac{1}{2}} \tan\left(\frac{E}{2}\right)
$$

Calculamos la anomalía excéntrica a partir de la fase orbital y luego la anomalía verdadera a partir de la anomalía excéntrica con la relación de las tangentes.
Podemos hacerlo iterativamente, creando un bucle con un comando "while"
y definiendo el grado de precisión que queremos en el valor de E



```python
def anomalia(fase, e):
    cota = 1.0e-6    # da la precisión para el cálculo de E
    nu   = 0.              # inicializa anomalía verdadera en cero radianes
    phi  = fase*2*np.pi  # convierte fase orbital a fase orbital angular (intervalo 0, 2*pi)
    E    = phi   # inicializa anom. exc. E
    dE   = 1.0   # inicializa delta de anom. exc. 
    while (np.sqrt(dE**2)>=cota):   
            dE = (e*np.sin(E) + phi - E) / ( 1. - e*np.cos(E))
            E  = E+dE
            
    nu = 2.*np.arctan2(np.tan(E/2.)*np.sqrt((1.+e)/(1.-e)),1.)  # calcular anomalía verdadera con ec. tangentes
    return nu
```

```python

# Definimos el parámetro e estimado asumiendo un error de un orden de magnitud menor
e,de = 0.2, 0.02


# Calulamos la anomalía verdadera para cada valor de fase
nu = np.array( [ anomalia(f, e) for f in fase] )

```

Dados $e$, $V_0$, $K$ y $\nu$ se utilizó la ecuación de velocidad radial para graficar curvas de velocidad radial teóricas para distintos valores de $\omega$ y compararlas con las observaciones de la componente primaria.


```python
# Definimos la función de la velocidad radial

def Vr(nu,K,V0,e,omega):
    return V0 + K * (e*np.cos(omega) + np.cos(omega + nu))

```

* Se calculó  el residuo $ ( O - C ) = V_r^{\text{observada}} - V_r^{\text{teórica}} $ en cada fase y su valor cuadrático medio (r.m.s) que se utilizará como una métrica para verificar si el residuo disminuye a medida que corregimos los parámetros orbitales. En esta primera estimación $ \text{r.m.s} = 6.2444 \, \text{km/s} $.


```python
#  pasamos la velocidad radial teórica a una variable fija 
Vr_t = Vr(nu,K,V0,e,omega)

residuo = data.VR1 - Vr_t
n = len(residuo)
rms = np.sqrt(np.sum(residuo**2)/n)
print('r.m.s. de (O-C) =', rms)
```

Utilizando $e = 0.02 \pm 0.02$, $V_{0} = (-10 \pm 5) \, \text{(km/s)}$, $K = 77.4 \pm 0.5$ y $\omega = 4 \pm 0.5$  se calcularon los valores numéricos de las derivadas parciales: 
$$
\frac{\partial V_r}{\partial K} = \cos(\nu + \omega) + e \cos(\omega)
$$

$$
\frac{\partial V_r}{\partial e} = K \left( \cos(\omega) - \frac{\sin(\nu + \omega) \sin(\nu)(2 + e \cos(\nu))}{(1 - e^2)} \right)
$$

$$
\frac{\partial V_r}{\partial \omega} = - K (\sin(\nu + \omega) + e \sin(\omega))
$$

y se ajustó por mínimos cuadrados la función $\Delta V_{r}$ a los datos de $(O-C)$.



```python

# Definimos la función Delta Vr en funcion de las derivadas parciales
# Asumimos que los ángulos vienen dados en radianes
# suponemos que P y T0 son fijos, por tanto sus derivadas son nulas y no las incluimos

#definimos los errores de los parámetros k, v0 y omega
dK,dV0,domega = 0.5, 5. , 0.5

def DeltaV(nu, dK, de, domega, dV0):
    # calculamos las derivadas parciales de f = DeltaV
    dfdK = np.cos( nu + omega) + e*np.cos(omega)
    dfde = K*((np.cos(omega)) - (np.sin( nu + omega)*np.sin(nu)*(2. + e*np.cos(nu)))/(1. - e**2.))
    dfdomega = - K*(np.sin( nu+ omega) + e*np.sin(omega))
    Delta_V = dfdK * dK + dfde * de + dfdomega * domega + dV0
    return Delta_V
```

```python
# Calculamos las variaciones en los parámetros orbitales con un ajuste de minimos cuadrados usando curve_fit 
# la función de ajuste es DeltaV, la variable independiente "nu" y los valores a ajustar son los residuos

deltas, cov = curve_fit(DeltaV, nu, residuo)
coef_err = np.sqrt(np.diag(cov))  ### los errores de los coeficientes ajustados (deltas)
                                  ### son la raíz cuadrada de los elementos de la diagonal

print('Deltas (correcciones) en K, e, omega, V0')
print(deltas)
print('Errores en Delta K, Delta e, Delta omega, Delta V0')
print(coef_err)
print('Matriz de covarianza:')
print(cov)
```

```python
# los nuevos valores son los viejos más la corrección
K_     = K     + deltas[0]
e_     = e     + deltas[1]
omega_ = omega + deltas[2]
V0_    = V0    + deltas[3]

print('K1, e1, omega1, gamma1', K_, e_, omega_, V0_)

```

```python
# volvemos a calcular los valores de anomalía verdadera y velocidad radial
# para cada fase con los parámetros corregidos

nu_ = np.array( [ anomalia(f, e_) for f in fase] )
Vr_t_ = Vr(nu_,K_,V0_,e_,omega_)

# y ahora calculamos un array con los nuevos residuos (O-C) y su r.m.s.
residuo_ = data.VR1 - Vr_t_

# Calculamos el error "medio". Esto es, el r.m.s. de los errores reportados por 
# los autores de las medidas. 
# Usamos esto para ver si el proceso ya nos da residuos menores que 
#los errores observacionales.

VR1_err_rms = np.sqrt(np.sum(data.VR1_err**2)/n)
print('r.m.s. de errores en VR primaria = ', VR1_err_rms)

# calculamos el r.m.s. de los nuevos (O-C)
rms_ = np.sqrt(np.sum(residuo_**2)/n)
print('r.m.s. de (O-C) corregidos =', rms_)
```

```python

fig = plt.figure(figsize=(10,5))

plt.errorbar(fase, data.VR1, yerr=data.VR1_err, fmt='.', markersize=10, capsize=5, label='$V_R$ Observación')
             

plt.scatter(fase,Vr_t,label='$V_R$ teórica orden 0', color= 'black' )

plt.scatter(fase,Vr_t_, label='$V_R$ teórica orden 1', color= 'red')


plt.title('Velocidades radiales de V501 Mon primaria observada y teórica orden 0 y 1')
plt.xlabel('fase $\phi$')
plt.ylabel(r'$V_r$ [km s$^{-1}$]')
plt.legend()


plt.savefig('orden0y1.png', dpi=100) 

# otro gráfico para ver cómo evoluciona la convergencia            
plt.figure(figsize=(10,5))
plt.scatter(np.sort(fase), abs(data.VR1_err), label='Error de obs')
plt.scatter(np.sort(fase),abs(residuo), lw=3,alpha=.8, label='Residuo orden 0')
plt.scatter(np.sort(fase),abs(residuo_), lw=3,alpha=.8, label='Residuo orden 1')
plt.legend()
plt.xlabel('fase $\phi$')
plt.ylabel(r'|O - C|')
plt.savefig('convergencia.png', dpi=100) 

## Observar si los residuos son aleatorios o si tienen "estructura" 
## Y si son del orden del error observacional o no.
```

Se observa que la curva de velocidad radial teórica calculada con los parámetros corregidos se ajusta en mayor medida a las observaciones.

* Iteramos los pasos anteriores hasta que el *r.m.s.* de las diferencias $(O-C)$ 
sea similar al *r.m.s.* de los errores en las mediciones de $V_{r,1}$ reportados por
Torres et al. (2015).

```python

# Pedimos que el módulo de la diferencia entre el rms de los residuos y el error medio
# asociado a los datos sea menor a 0.005

tolerancia = 0.005
lim= 100.  # nro. límite de iteraciones
K_,e_,omega_,V0_ = K,e,omega,V0   # valor inicial de los parámetros

i=0  # contador de iteraciones
flag= True

while flag:
    i+=1
    K_     = K_     + deltas[0]
    e_     = e_     + deltas[1]
    omega_ = omega_ + deltas[2]
    V0_    = V0_    + deltas[3]

    nu_ = np.array( [ anomalia(f, e_) for f in fase] )

    Vr_t_=Vr(nu_,K_,V0_,e_,omega_)

    residuo_ = data.VR1 - Vr_t_

    rms_ = np.sqrt(np.sum(residuo_**2)/n)
    
    criterio = (rms_- VR1_err_rms ) < tolerancia 
    
    if ((i>=lim) | criterio) :
        flag = False
    else:
        deltas, cov = curve_fit(DeltaV, nu_, residuo_)
        coef_err = np.sqrt(np.diag(cov)) 
        print('Iteraciones:',i)
        print('r.m.s. de los residuos (O-C) =',rms_)
      
```

Sabemos que la velocidad radial de cada componente del sistema está dada por:

$$
V_{R1} = V_{0} + K_{1}[ecos(\omega_{1}) + cos(\omega_{1} + \nu)]
$$

$$
V_{R2} = V_{0} + K_{2}[ecos(\omega_{2}) + cos(\omega_{2} + \nu)]
$$

y que se encuentran a contrafase, es decir, $\omega_{2} = \omega_{1} + 180^{\circ}$. Podemos expresar $V_{R1}$ en función de $V_{R2}$.

$$
V_{R1} = -qV_{R2} + V_{0}(1+q)
$$

donde $q = \frac{K_{1}}{K_{2}}$.

Usando que la relación entre las velocidades radiales de las componentes es lineal con pendiente $-q$, graficamos $V_{R1}$ vs $-V_{R2}$ y realizamos un ajuste lineal para obtener el cociente de masas $q$.



```python
def f_q(x, m, n):
    'función lineal'
    f_q = m * x + n
    return f_q 

coef_q, cov_q = curve_fit(f_q,-data.VR2 , data.VR1)
coef_err_q = np.sqrt(np.diag(cov_q))  

print('Coeficientes recta m,n')
print(coef_q)
print('Errores en m y n:')
print(coef_err_q)
print('Matriz de covarianza:')
print(cov_q)
#print('por tanto q =', m)
```
